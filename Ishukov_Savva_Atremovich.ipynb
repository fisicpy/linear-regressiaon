В файле travel_time_data.csv содаржатся данные о времени поездки (в часах) в зависимости от расстояния до пункта назначения (в километрах). Нобходимо решить задачу линейной регрессии: создать и обучить модель линейной регрессии, которая будет способна прогнозировать время поездки по известному расстоянию до пункта назначения.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as mse

travel_time_train_data = pd.read_csv("travel_time_train_data.csv")
print(travel_time_train_data.head())

x = travel_time_train_data["distance"]
y = travel_time_train_data["travel_time"]
plt.scatter(x, y)
plt.ylim(0, 30)
plt.xlabel("distance")
plt.ylabel("travel_time")
plt.show()

train_data, test_data = train_test_split(
	travel_time_train_data,
	test_size=0.4,
	random_state=42
)

x_train = train_data["distance"].to_numpy()
y_train = train_data["travel_time"].to_numpy()

x_test = test_data["distance"].to_numpy()
y_test = test_data["travel_time"].to_numpy()

print(f"{x_train.shape = }, {y_train.shape = }")
print(f"{x_test.shape = }")

x_train = x_train.reshape((63, 1))
x_test = x_test.reshape((43, 1))

print(f"{x_train.shape = }, {y_train.shape = }")
print(f"{x_test.shape = }")

model = LinearRegression()
model.fit(x_train, y_train)

y_test_predicted = model.predict(x_test)

header = f"X_test\ty_test\ty_test_predicted"
print(header, '\n', '-'*len(header), sep='')
for i in range(len(y_test)):
    print(f"{x_test[i]}\t{y_test[i]}\t{int(y_test_predicted[i])}")

y_train_predicted = model.predict(x_train)

plt.scatter(x_train, y_train, c='grey', s=15)
plt.scatter(x_test, y_test, c='green')
plt.plot(x_train, y_train_predicted)
plt.ylim(0, 30)
plt.xlabel('distance')
plt.ylabel('travel_time')
plt.show()


train_rmse = mse(y_train, y_train_predicted, squared=False)
test_rmse = mse(y_test, y_test_predicted, squared=False)

print(f'{train_rmse = }, {test_rmse = }')

travel_time_test_data = pd.read_csv("travel_time_test.csv")
print(travel_time_test_data.head())

x_test_result = travel_time_test_data.to_numpy()
y_test_predicted_result = model.predict(x_test_result)
print(y_test_predicted_result[:5])

np.savetxt(
    "Ishukov_Savva_Artemovich_travel_time_predict.csv",
    y_test_predicted_result,
    header="travel_time_predict",
    delimiter=",",
    fmt="%.2f",
    comments=""
)
```
